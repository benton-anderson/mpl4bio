## Motivation

Biology research can be broadly divided into two groups:
    1. Field and lab work directly on living things
    2. Analyzing the molecules that make biological systems so complex

The second is a much more modern undertaking, and can be referred to as molecular biology or computational biology, among other names.  

This field is extremely diverse, pulling in researchers from all scientific fields from mathematics to chemistry to psychology, and even physicians, ethicists and lawyers. 

A high-impact biology paper can draw on numerous avenues of analysis, for example: 
1. Genomics and genome-wide association studies (GWAS)
2. single cell RNA sequencing (scRNAseq) 
3. Chromatin accessibility (atacseq) 
4. Mass spectrometry proteomics and metabolomics to analyze proteins and small molecules 

The common thread between all of these fields is the -omics suffix, which means ( borrowing from wikipedia)
> Broad characterization and quantification of large groups of biological molecules that translate into the structure, function, and dynamics of a biological system. 

And all of these fields have very large datasets. And each field has their own set of techniques, data sets, data types, and from that, a very wide variety of possible visualizations that enable communication of the complexity of the research findings in digestable, picture-based format. 

Browsing through almost any modern biology paper that uses an -omics technique will reveal Figures, divided into subfigures, AKA panels, and each will have respective axes, legends, colors, shapes, lines, anything. Common examples of -omics figures are the Volcano plot, to quantify the difference and significance between two research conditions; and UMAP, which is often used to distill massive datasets (think thousands of rows and ten thousand columns of nothing but numbers) into a 2-dimensional representation of how your research observations relate to each other.

## Basic building blocks for beautiful figures

The basics for making effective figures in Python starts with knowledge of the most fundamental Python plotting package: Matplotlib. The library has its roots in the plotting library from MATLAB, yet with these ancient roots it has become the ultimate package for customizing your figure exactly how you want. Unfortunately, this extensive customization ability over the years has led to feature bloat, some API choices that seem odd at first glance, and two different ways of interacting with your plots that can lead to no end of confusion (for example, see the 70k stackoverflow questions on MPL) 

I will start by introducing the two separate methods for interacting with figures:
1. the state-machine approach
    - Anything that starts with `plt.` is in this category
2. The object-oriented approach
    - Generally, anything with `fig.` or `.ax` fits here
    
I have learned that it is by far the best approach to start every new figure with exactly ONE LINE of the state machine `plt.` code, then never touch it again. The starting point is: 
`fig, ax = plt.subplots()`

Notice how I pull out the `fig` and `ax` object-oriented objects from the `plt.subplots()`

`fig` is the entire figure, and there is only one figure per visualization. 
`ax` is just one `axes` object, and there can be multiple `axes` within a `fig`, and as we saw in the examples of complex biology figures above, we will often use multiple `axes` objects per figure. 

To better illustrate `fig` and `ax`, here is the official matplotlib annotated guide to the objects within a figure:


Notice that `axes` is not the x- and y-axis, but rather the entire area on which your content is plotted. All of the sub-objects of `ax` are plotted onto your `fig` which is the "canvas" that `matplotlib.artists` paint onto (get it?)

And `plt.subplots()` is not limited to just the default parameters. You can also do 
`fig, axs = plt.subplots(ncols=1, nrows=2)` where `axs` is a list of two `axes` objects
Alternatively, you can use Python's tuple unpacking on the list of `axes` like so `fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(ncols=2, nrows=2)`

But at any rate, this is the number one best way to enter into plotting with matplotlib and I strongly recommend you use it every time. 


## Doing something with your `ax` objects

Now that you have an ax object, the next step is to plot something on it. It could be simple like `ax.bar` for a barplot, or `ax.scatter` or `ax.plot` for a line plot. 

However, these default styles are quite rough and often insufficient for conveying the complexity of your awesome research. 

Instead, almost every single pure Matplotlib plot type (bar, scatter, line, etc.) has an associated `seaborn` plot
`import seaborn as sns
sns.scatterplot([1,2,3], [3,2,1])`

And all* seaborn plots have the option to add an `ax=ax` parameter so that the seaborn plot goes directly into the `ax` objects you made with your previous call to `plt.subplots()`

for example:



*except for a few seaborn plots known as "Figure-level plots", such as Clustermaps, lmplot and FacetGrids, that we will come to later

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)
sns.scatterplot(x=[1,2,3], y=[3,2,1], ax=ax1)
sns.barplot(x=[0,1,2], y=[5, 3, 6], ax=ax2)
sns.lineplot(x=[1,2,3], y=[1, 4, 2], ax=ax3)
sns.histplot(data=[2,3,4,5,1,5,7,4,2,4,7,8,], ax=ax4)
sns.despine()
```

There! Isn't that quite beautiful already? And only 5 lines of code (I use `sns.despine()` at the bottom of almost every plot I make to remove the upper and right-side spines to give the plots a much cleaner appearance). 

The last trick for basic-level plot improvements involves using a few more parameters in `plt.subplots()`.

There are many useful parameters in plt.subplots(), and even for super fast, small plots that I use only for my own quick checks, I include two extra parameters:
`plt.subplots(dpi=300, figsize=(5, 5))`
`dpi` refers to "Dots per inch" (sorry metric users), and `figsize` gives the figure size in inches.

The boost in size and DPI makes your figure much cleaner, and ensures that if you copy/paste from your browser into PowerPoint, you'll get a higher quality, less fuzzy image. 

It's worth noting that the popular `pandas` library includes a `.plot()` method for `Series` and `DataFrame` objects. The `.plot()` method also accepts the `ax=` keyword `df.plot(kind='line', ax=ax1)` and this can be used to plot directly from dataframes into your `axes` like with seaborn. 

# Basics, tl;dr

# Intermediate-level plotting 

Knowing these basics of figures and axes, and using seaborn as your go-to plotting library will get you far, until you start to think about a few fundamentals of good plots:
1. Clearly labeled x-axis 
2. Clearly labeled y-axis
2. Descriptive title
3. Legend (sometimes needed, sometimes not)

Notice that our plots above are lacking all of these elements. 

### Omitting the first 3 labels is bad, and their inclusion is non-negotiable.
#### Legends are optional depending on your figure

To add these elements, we use the matplotlib object-oriented calls within each `axes` and sometimes within `fig`



```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, gridspec_kw=dict(wspace=0.5, hspace=0.5))
sns.scatterplot(x=[1,2,3], y=[3,2,1], ax=ax1)
sns.barplot(x=[0,1,2], y=[5, 3, 6], ax=ax2, label='Legend label')
sns.lineplot(x=[1,2,3], y=[1, 4, 2], ax=ax3)
sns.histplot(data=[2,3,4,5,1,5,7,4,2,4,7,8,], ax=ax4)
ax1.set_title('Fancy dots')
ax1.set_xlabel('First dimension')
ax1.set_ylabel('Really cool result')



# Setting each title and label individually allows for customization of fontsizes
ax2.set_title('Big bars, big font', fontsize=18)
ax2.set_xlabel('Small fontsize', fontsize=6)
ax2.set_ylabel('Just-right fontsize', fontsize=10)
# Not all plots need legends, but this one would do well 
ax2.legend(loc=(1, 0))
ax3.set_title('My line plot')
ax3.set_xlabel('Bold titles are bold', fontweight='bold')
ax3.set_ylabel('Multi-line labels\nare cool')
# You can also use the generic `.set()` method to add all at once, with less customization of font properties
ax4.set(title='Check this histogram', xlabel='My category', ylabel='Money')
sns.despine()
```

Notice how I added `gridspec_kw=dict(wspace=0.5, hspace=0.5)` to `plt.subplots()`. This affects the matplotlib "Grid Specification" to add more vertical and horizontal spacing between subplots. Without it, the plots overlap and the titles and labels get cut off, try it without the `gridspec_kw` to see for yourself!
